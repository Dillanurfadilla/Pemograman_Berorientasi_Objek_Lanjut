```python
class Music:
   def __init__(self, nama, tahun_rilis):
       self.nama = nama
       self.tahun_rilis = tahun_rilis 
   
   def genre(self):
       print(f"{self.nama} bergenre disko_pop memuncaki tangga lagu billboard.")
       
class Dynamite(Music):
   def __init__(self, nama, negara, penyanyi):
       super().__init__(nama, negara)
       self.penyanyi = penyanyi 
  
   def penghargaan(self):
       print(f"{self.nama} lagu dari {self.penyanyi} memenangkan 21 penghargaan ditahun 2020.")
    
DynamiteA = Dynamite("Dynamite", 2020, "BTS")
DynamiteA.genre() # Output: DynamiteAgenre disko_pop memuncaki tangga lagu billboard.
DynamiteA.penghargaan() # Output: Dyamite lagu dari BTS memenangkan 21 penghargaan di tahun 2020.
```
