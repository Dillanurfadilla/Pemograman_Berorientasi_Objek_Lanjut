```python
class Merek_kendaraan:
   def __init__(self, name):
       self.name = name
       
   def darat(self):
       print(f"{self.name} darat")
       
class Honda(Merek_kendaraan):
   def __init__(self, name, two_wheeled):
       super().__init__(name)
       self.two_wheeled = two_wheeled 
       
   def petrol(self):
        print(f"{self.name} fuel petrol with a two_wheeled  {self.two_wheeled}")
        
class Motor(Honda):
   def __init__(self, name, two_wheeled, color):
       super().__init__(name, two_wheeled)
       self.color = color
       
   def matic(self):
       print(f"{self.name} is a {self.color} Beat that matic")
       
Motor = Motor ("Beat", " two_wheeled", "red")
Motor.matic() # Output: Beat is a red Motor that matic
Motor.petrol() # Output: Beat fuel petrol with a two_wheeled of red
```
