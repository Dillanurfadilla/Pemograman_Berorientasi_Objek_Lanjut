```python

class Movie:
   def __init__(self, nama, tahun_rilis):
       self.nama = nama
       self.tahun_rilis = tahun_rilis 
   
   def genre(self):
       print(f"{self.nama} bergenre Fantasi.")
       
class Harry_Potter(Movie):
   def __init__(self, nama, tahun_rilis, penulis):
       super().__init__(nama,tahun_rilis)
       self.penulis = penulis 
  
   def berasal(self):
    print(f"{self.nama} karya {self.penulis} berasal dari inggris.")
    
Harry_PotterA = Harry_Potter("Harry Potter", 2001, "jk.Rowling")
Harry_PotterA.genre() # Output: Harry_Potter  genre Fantasi.
Harry_PotterA.berasal() # Output: Harry_Potter karya jk.rowling berasal dari inggris.
```
