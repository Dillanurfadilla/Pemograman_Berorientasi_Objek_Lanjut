```python
class buah:
   def __init__(self, nama, kandungan_vitamin):
       self.nama = nama
       self.kandungan_vitamin  = kandungan_vitamin 
       
   def display_info(self):
        print(f"Nama: {self.nama}")
        print(f"kandungan_vitamin: {self.kandungan_vitamin}")
        
class panas:
   def __init__(self, jenis, iklim):
       self.jenis = jenis
       self.iklim = iklim      
   def display_info(self):
        print(f"jenis: {self.jenis}")
        print(f"iklim: {self.iklim }")
        
class tropis:
   def __init__(self, iklim , rasa):
       self.ikl = iklim
       self.rasa = rasa
       
   def display_info(self):
       print(f"iklim: {self.iklim}")
       print(f"rasa: {self.rasa}")
       
class Apel(panas, tropis):
   def __init__(self, nama, kandungan_vitamin, jenis, iklim, rasa):
       buah.__init__(self, nama, kandungan_vitamin)
       panas.__init__(self, jenis, iklim)
       tropis.__init__(self, iklim, rasa)
     
   def display_info(self):
        super().display_info()
        print(f"Nama: {self.nama}")
        print(f"kandungan_vitamin: {self.kandungan_vitamin}")
        print(f"jenis: {self.jenis}")
        print(f"iklim: {self.iklim}")
        print(f"rasa:{self.rasa}")
        
# contoh penggunaan
ApelA = Apel("Apel", "C", "Berdaging", "Subtropis", "Manis")
ApelA.display_info()
```
